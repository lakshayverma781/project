```python
import pandas as pd
df=pd.DataFrame(pd.date_range('1/1/2021','1/31/2021'),columns=['date'])
df['year']=df['date'].dt.year
df['quarer']=df['date'].dt.quarter
df['month']=df['date'].dt.month
df['day']=df['date'].dt.day
df['weekday']=df['date'].dt.weekday
df['month_name']=df['date'].dt.month_name(locale='English')
df['week_name']=df['date'].dt.day_name(locale='English')
df
```

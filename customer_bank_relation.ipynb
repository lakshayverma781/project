```python
#  Print the first 10 fibonacci number 
n = 10
n1 = 0
n2 = 1
count = 0
if n<= 0:
   print('enter a positive integer')
elif n == 1:
   print(f'fibonacci series upto {n}:')
   print(n1)
else:
   print('Fibonacci series:')
   while count < n:
       print(n1,end='  ')
       nth=n1+n2
       n1 = n2
       n2 = nth
       count += 1
```
